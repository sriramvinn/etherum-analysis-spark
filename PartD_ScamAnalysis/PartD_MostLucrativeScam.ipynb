```python
import pandas as pd
#import matplotlib as plt
from matplotlib import pyplot as plt
import seaborn as sb
```

```python
%config InlineBackend.figure_formats = set(['retina'])
sb.set_style('darkgrid')
```

```python
df = pd.read_csv('monthly_scam_transactions.csv', names=['Type', 'Month', 'TotalEther', 'NumTransactions', 'AverageEther'])
```

```python
df['Month'] = pd.Categorical(df['Month'], ordered=True)
```

```python
df['Year'] = df.apply(lambda row: row['Month'][:4], axis=1)
```

```python
years = df['Year'].unique()
```

```python
fig, ax = plt.subplots(nrows = len(years))
fig.set_figheight(15)
fig.set_figwidth(15)
i = 0
for year in sorted(years):
    sb.lineplot(data=df[df['Year'] == year], x='Month', y='TotalEther', hue='Type', ax=ax[i])
    i = i + 1
```

```python
fig2, ax2 = plt.subplots(nrows = len(years))
fig2.set_figheight(15)
fig2.set_figwidth(15)

i = 0
for year in sorted(years):
    sb.lineplot(data=df[df['Year'] == year], x='Month', y='AverageEther', hue='Type', ax=ax2[i])
    i = i + 1
```

```python
fig3, ax3 = plt.subplots(nrows = len(years))
fig3.set_figheight(15)
fig3.set_figwidth(15)


i = 0
for year in sorted(years):
    sb.lineplot(data=df[df['Year'] == year], x='Month', y='NumTransactions', hue='Type', ax=ax3[i])
    i = i + 1
```

```python
fig4, ax4 = plt.subplots(nrows = len(years), ncols=3)
fig4.set_figheight(15)
fig4.set_figwidth(15)


i = 0
for year in sorted(years):
    sb.lineplot(data=df[df['Year'] == year], x='Month', y='TotalEther', hue='Type', ax=ax4[i][0]).set_title('Total Ether received by each type of scam in {0}'.format(year))
    sb.lineplot(data=df[df['Year'] == year], x='Month', y='AverageEther', hue='Type', ax=ax4[i][1]).set_title('Average Ether received by each type of scam in {0}'.format(year))
    sb.lineplot(data=df[df['Year'] == year], x='Month', y='NumTransactions', hue='Type', ax=ax4[i][2]).set_title('Num of transactions by each type of scam in {0}'.format(year))
    i = i + 1
```
